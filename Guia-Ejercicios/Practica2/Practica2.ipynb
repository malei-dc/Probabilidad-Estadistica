### (a) Graficar funci√≥n de probabilidad puntual asociada a X 

```python
import matplotlib.pyplot as plt

# Valores de X y sus probabilidades correspondientes
X = [0, 1, 2, 3]
probabilidades = [0.3943, 0.4302, 0.1563, 0.0189]

# Graficar la funci√≥n de probabilidad
plt.bar(X, probabilidades, color='skyblue')
plt.xlabel('N√∫mero de art√≠culos defectuosos (X)')
plt.ylabel('Probabilidad')
plt.title('Funci√≥n de probabilidad de X (n√∫mero de defectuosos)')
plt.xticks(X)  # Mostrar s√≥lo los valores de X en el eje x
plt.show()
```

### (c) Graficar funci√≥n de distribuci√≥n acumulada de X 

```python
import matplotlib.pyplot as plt
import numpy as np

# Valores de X y sus probabilidades correspondientes
X = [0, 1, 2, 3]
probabilidades = [0.3943, 0.4302, 0.1563, 0.0189]

# Calcular la funci√≥n de distribuci√≥n acumulada (CDF)
cdf = np.cumsum(probabilidades)

# Graficar la CDF
plt.step(X, cdf, where='post', color='purple')
plt.fill_between(X, cdf, step="post", alpha=0.4, color='purple')
plt.xlabel('N√∫mero de art√≠culos defectuosos (X)')
plt.ylabel('Probabilidad acumulada')
plt.title('Funci√≥n de distribuci√≥n acumulada (CDF) de X')
plt.xticks(X)
plt.ylim(0, 1)  # La CDF va de 0 a 1
plt.show()
```

### (d) Estimar mediante una simulaci√≥n las probabilidades calculadas en el item (a).

```python
import numpy as np
import matplotlib.pyplot as plt

simulaciones = 100000

defectuosos = 4
no_defectuosos = 15 - defectuosos

# Simulaci√≥n de selecciones (con reposici√≥n)
resultados = np.random.choice([0, 1], size=(simulaciones, 3), p=[no_defectuosos/15, defectuosos/15])

# Sumar cu√°ntos art√≠culos defectuosos se seleccionan en cada experimento
num_defectuosos = np.sum(resultados, axis=1)

# Calcular la frecuencia relativa de cada posible valor de X (0, 1, 2, 3)
valores, frecuencias = np.unique(num_defectuosos, return_counts=True)
probabilidades_simuladas = frecuencias / simulaciones

print(f"Probabilidades estimadas para X = 0, 1, 2, 3: {probabilidades_simuladas}")
print(f"Probabilidades calculadas para X = 0, 1, 2, 3: [0.3943 0.4302 0.1563 0.0189]")

# Graficamos
plt.bar(valores, probabilidades_simuladas, color='lightgreen')
plt.xlabel('N√∫mero de art√≠culos defectuosos (X)')
plt.ylabel('Probabilidad estimada')
plt.title('Probabilidades estimadas mediante simulaci√≥n')
plt.xticks(valores)
plt.show()
```

### (d) Mediante una simulaci√≥n, estimar las probabilidades del item b).

```python
import random

valores_X = [1, 3, 6, 12]
probabilidades_X = [0.3, 0.1, 0.2, 0.4]

n_simulaciones = 10000
realizaciones = random.choices(valores_X, probabilidades_X, k=n_simulaciones)

# Estimar P(3 < X <= 6)
p_3_menor_X_leq_6 = sum(3 < x <= 6 for x in realizaciones) / n_simulaciones

# Estimar P(3 <= X <= 6)
p_3_leq_X_leq_6 = sum(3 <= x <= 6 for x in realizaciones) / n_simulaciones

# Estimar P(X >= 4)
p_X_geq_4 = sum(x >= 4 for x in realizaciones) / n_simulaciones

# Estimar P(X >= 6)
p_X_geq_6 = sum(x >= 6 for x in realizaciones) / n_simulaciones

# Mostrar resultados
print(f"P(3 < X <= 6): {p_3_menor_X_leq_6}")
print(f"P(3 <= X <= 6): {p_3_leq_X_leq_6}")
print(f"P(X >= 4): {p_X_geq_4}")
print(f"P(X >= 6): {p_X_geq_6}")
```

# Ejercicio 3



```python
# Definir la funci√≥n F_X
def F_X(x, p, t):
    acumulado = 0
    
    # Recorrer los valores de x
    for i in range(len(x)):
        if x[i] <= t:
            acumulado += p[i]
    
    return acumulado

# Probar la funci√≥n con los datos del ejercicio anterior
x = [1, 3, 6, 12]  # Valores posibles de X
p = [0.3, 0.1, 0.2, 0.4]  # Probabilidades correspondientes

# Calcular F_X(t) para diferentes valores de t
print(F_X(x, p, 0))   # Deber√≠a devolver 0
print(F_X(x, p, 1))   # Deber√≠a devolver 0.3
print(F_X(x, p, 3))   # Deber√≠a devolver 0.4
print(F_X(x, p, 6))   # Deber√≠a devolver 0.6
print(F_X(x, p, 12))  # Deber√≠a devolver 1
```

# Ejercicio 6

### Funci√≥n que calcula esperanza

```python
import numpy as np

def calcular_esperanza(x, p):
    # Verificar que los vectores tengan la misma longitud
    if len(x) != len(p):
        raise ValueError("Los vectores 'x' y 'p' deben tener la misma longitud.")
    
    # Verificar que las probabilidades sumen 1
    if not np.isclose(sum(p), 1):
        raise ValueError("Las probabilidades en 'p' deben sumar 1.")
    
    # Calcular la esperanza
    esperanza = np.sum(np.array(x) * np.array(p))
    return esperanza

# Prueba de la funci√≥n con la variable aleatoria X del ejercicio 1
x = [0, 1, 2, 3]  # Valores de X
p = [0.3943, 0.4302, 0.1563, 0.0192]   # Probabilidades asociadas

# Calcular la esperanza
esperanza = calcular_esperanza(x, p)
print(f"La esperanza de X es: {esperanza}")
```

# Ejercicio 7

### Funci√≥n que calcula la varianza.

Teniendo en cuenta que la $Var(X)$ se define como: 

$$Var(X) = E(X^2) + (E(X))^2$$

```python
import numpy as np

def calcular_varianza(x, p):
    # Verificar que los vectores tengan la misma longitud
    if len(x) != len(p):
        raise ValueError("Los vectores 'x' y 'p' deben tener la misma longitud.")
    
    # Verificar que las probabilidades sumen 1
    if not np.isclose(sum(p), 1):
        raise ValueError("Las probabilidades en 'p' deben sumar 1.")
    
    # Calcular la esperanza (media)
    esperanza = np.sum(np.array(x) * np.array(p))
    print(f"La esperanza de X es: {esperanza}")
    
    # Calcular la esperanza de X^2
    esperanza_cuadrado = np.sum(np.array(x)**2 * np.array(p))
    
    # Calcular la varianza: Var(X) = E[X^2] - (E[X])^2
    varianza = esperanza_cuadrado - esperanza**2
    return varianza

# Prueba de la funci√≥n con la variable aleatoria X del ejercicio 1
x = [0, 1, 2, 3]  # Valores de X
p = [0.3943, 0.4302, 0.1563, 0.0192]  # Probabilidades asociadas

# Calcular la varianza
varianza = calcular_varianza(x, p)
print(f"La varianza de X es: {varianza}")
```
