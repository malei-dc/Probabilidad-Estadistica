```python
import numpy as np
import matplotlib.pyplot as plt

# Valores de X y sus probabilidades acumuladas
X = [-20000, -10000, 0, 10000, 20000]
F = [0.0545, 0.2181, 0.5453, 0.818, 1]

# Graficar la función de distribución acumulada
plt.step(X, F, where='post', label='F(x)', color='blue')

# Etiquetas y título
plt.xlabel('X')
plt.ylabel('F(x)')
plt.title('Función de Distribución Acumulada de X')
plt.grid(True)
plt.show()
```
