# Ejercicio 1

### (d) Graficar $f_X$ y $F_X$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir las funciones f(x) y F(x)
def f(x):
    return 0.75 * (1 - x**2)

def F(x):
    return 0.75 * (x - x**3 / 3 + 2/3)

# Generar datos de x en el intervalo [-1, 1]
x_vals = np.linspace(-1, 1, 500)

# Evaluar las funciones f(x) y F(x)
f_vals = f(x_vals)
F_vals = F(x_vals)

# Crear las gráficas
fig, axs = plt.subplots(2, 1, figsize=(8, 10))

# Graficar f(x)
axs[0].plot(x_vals, f_vals, label='$f(x)$', color='blue')
axs[0].fill_between(x_vals, f_vals, color='lightblue', alpha=0.5)
axs[0].set_xlim([-1, 1])
axs[0].set_ylabel('f(x)')
axs[0].set_title('Función de densidad $f(x)$')
axs[0].grid(True)

# Graficar F(x)
axs[1].plot(x_vals, F_vals, label='$F(x)$', color='green')
axs[1].fill_between(x_vals, F_vals, color='lightgreen', alpha=0.5)
axs[1].set_xlim([-1, 1])
axs[1].set_ylabel('F(x)')
axs[1].set_title('Función de distribución $F(x)$')
axs[1].grid(True)

# Mostrar gráficos
plt.tight_layout()
plt.show()
```

# Ejercicio 2

### (e) Graficar $f_X$ y $F_X$ . Agregar la recta vertical $x = \tilde{\mu} $.


```python
import numpy as np
import matplotlib.pyplot as plt

# Definir las funciones f_X(x) y F_X(x)
def f_X(x):
    return np.where((x >= 0) & (x <= 2), (3 * x**2) / 8, 0)

def F_X(x):
    return np.where(x < 0, 0, np.where(x <= 2, (x**3) / 8, 1))

# Valor de la mediana, que es x donde F_X(x) = 0.5
mu_tilde = np.cbrt(4)  # Aproximadamente 1.5874

# Generar datos de x en el intervalo [-1, 3] para tener un buen rango
x_vals = np.linspace(-1, 3, 500)

# Evaluar las funciones f_X(x) y F_X(x)
f_vals = f_X(x_vals)
F_vals = F_X(x_vals)

# Crear las gráficas
fig, axs = plt.subplots(2, 1, figsize=(8, 10))

# Graficar f_X(x)
axs[0].plot(x_vals, f_vals, label=r'$f_X(x)$', color='blue')
axs[0].fill_between(x_vals, f_vals, color='lightblue', alpha=0.5)
axs[0].axvline(x=mu_tilde, color='red', linestyle='--', label=r'$\tilde{\mu}$ (mediana)')
axs[0].set_xlim([-1, 3])
axs[0].set_ylim([0, 1.75])
axs[0].set_ylabel(r'$f_X(x)$')
axs[0].set_title(r'Función de densidad $f_X(x)$')
axs[0].grid(True)
axs[0].legend()

# Graficar F_X(x)
axs[1].plot(x_vals, F_vals, label=r'$F_X(x)$', color='green')
axs[1].fill_between(x_vals, F_vals, color='lightgreen', alpha=0.5)
axs[1].axvline(x=mu_tilde, color='red', linestyle='--', label=r'$\tilde{\mu}$ (mediana)')
axs[1].set_xlim([-1, 3])
axs[1].set_ylim([0, 1.2])
axs[1].set_ylabel(r'$F_X(x)$')
axs[1].set_title(r'Función de distribución $F_X(x)$')
axs[1].grid(True)
axs[1].legend()

# Mostrar gráficos
plt.tight_layout()
plt.show()
```
